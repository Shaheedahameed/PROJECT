# 1. Overview of Problem Statement: 

The Online Shoppers Purchasing Intention Dataset is a collection of data that is primarily concerned with understanding the factors that influence whether online shoppers make a purchase or not. The dataset comes from a simulation of user interactions on an e-commerce website. The central problem is to predict whether a visitor to the website will ultimately make a purchase, based on various attributes such as their behavior on the website, demographic information, and other contextual details. This problem is of significant interest for businesses looking to optimize their online marketing and improve conversion rates by tailoring their approach based on shopper behaviors.

The problem also revolves around identifying key factors that influence purchasing decisions in an online setting. The dataset includes a range of features like the time spent on the website, the number of pages viewed, whether the shopper is a returning visitor, and which marketing channels brought them to the website. Additionally, demographic features such as the user's region or device type are included to understand if these characteristics play a role in a shopper's intent to purchase. Analyzing these features could help marketers design more targeted strategies that cater to different segments of online shoppers, potentially increasing sales and improving the customer experience.

Machine learning algorithms, such as classification models, can be applied to this dataset to predict purchasing intentions. With this predictive modeling, businesses can gain insights into which factors are most predictive of a purchase, enabling them to make data-driven decisions. This could involve personalizing recommendations, optimizing website layout, or targeting users with specific ads. Ultimately, the goal is to minimize cart abandonment and increase the likelihood of a visitor converting into a customer.

### Features : 
#####  1 - Administrative : Number of pages visited in the "Administrative" section.
#####  2 - Administrative_Duration : Time spent (in seconds) on administrative pages.
#####  3 - Informational : Number of pages visited in the "Informational" section.
#####  4 - Informational_Duration : Time spent on the "Informational" section (in seconds).
#####  5 - ProductRelated : Number of pages visited in the "ProductRelated" section.
#####  6 - ProductRelated_Duration : Time spent on the "ProductRelated" section (in seconds).
#####  7 - BounceRates : Percentage of visitors who leave the site after viewing only one page.
#####  8 - ExitRates : Percentage of visitors who leave the site after viewing the current page.
#####  9 - PageValues : Value associated with each page viewed, based on its importance.
##### 10 - SpecialDay : A binary indicator of whether the visit occurred on a special day (e.g., a holiday or promotional event).
##### 11 - Month: The month of the session.
##### 12 - OperatingSystems : The operating system used by the visitor (e.g., Windows, Mac).
##### 13 - Browser : The browser used by the visitor (e.g., Chrome, Firefox).
##### 14 - Region : The geographic region from which the visitor accessed the website.
##### 15 - TrafficType : The type of traffic source that brought the visitor to the website (e.g., search engine, direct).
##### 16 - VisitorType : Whether the visitor is a "NewVisitor," "ReturningVisitor," or "Other".
##### 17 - Weekend : A binary indicator showing whether the visit occurred on a weekend or weekday.
##### 18 - Revenue : The target variable indicating whether the visitor made a purchase (1 = purchase, 0 = no purchase).

### Datecting and Treating Outliers using IQR Method

```python
# Function to cap outliers using quantiles
def cap_outliers(df, column, lower_percentile=5, upper_percentile=95):
    lower_cap = np.percentile(df[column], lower_percentile)
    upper_cap = np.percentile(df[column], upper_percentile)
    df[column] = np.clip(df[column], lower_cap, upper_cap)
    return df

# Apply capping to all numeric columns
for column in numeric_columns:
    data = cap_outliers(data, column)

# Visualize the data after capping the outliers
for column in numeric_columns:
    plt.figure(figsize=(12, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(data[column], kde=True)
    plt.title(f'Histogram of {column} after capping outliers')

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[column])
    plt.title(f'Boxplot of {column} after capping outliers')

    plt.show()
```

### Model Development and Performance Evaluation

```python
# Define a dictionary containing different regression models
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import accuracy_score,precision_score,f1_score,classification_report,confusion_matrix,recall_score,roc_curve, roc_auc_score

models = {
    '1.LinearRegression' : LinearRegression(),
    '2.DecisionTreeRegressor': DecisionTreeRegressor(),
    '3.SVR': SVR(),
    '4.RandomForestRegressor': RandomForestRegressor(),
    '5.GradientBoostingRegressor': GradientBoostingRegressor(),
    '6.MLPRegressor': MLPRegressor(),
    '7.AdaBoostRegressor' : AdaBoostRegressor()
}
```

```python
# Initialize results dictionary
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score, classification_report, confusion_matrix

results = {}

# Step 2: Train, predict, and evaluate each model
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Calculate performance metrics
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
    # Store results
    results[model_name] = {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R2': r2}
```

```python
 #Convert the results dictionary into a Pandas DataFrame
results_df = pd.DataFrame(results).T  # Transpose (T) to make models as rows and metrics as columns

# Display the evaluation results for all models
print(results_df)
```

```python
# Sort models by R² score in descending order and Extract the best-performing model (first row after sorting)
best_model = results_df.sort_values(by= 'R2',ascending = False).iloc[0]

# Display the best model name and its evaluation metrics
print("Best model is")
print(best_model)
```

```python
# Sort models by R² score in Ascending order and Extract the worst-performing model (first row after sorting)
worst_model = results_df.sort_values(by= 'R2',ascending = True).iloc[0]

# Display the best model name and its evaluation metrics
print("Worst model is")
print(worst_model)
```
